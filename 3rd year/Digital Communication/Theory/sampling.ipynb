```python
import numpy as np
import matplotlib.pyplot as plt

def continuous_signal(t, f0):
    return np.cos(2 * np.pi * f0 * t)

def undersampled_signal(n, f0, fs):
    Ts = 1 / fs
    sampled_values = continuous_signal(n * Ts, f0)
    return sampled_values

# Signal parameters
f0 = 1  # Frequency of the original signal
fs = 1.5 * f0  # Sampling rate

# Time vector for continuous signal
t_continuous = np.linspace(0, 4, 1000, endpoint=False)

# Time vector for undersampled signal
n_undersampled = np.arange(0, 4, 1 / fs)

# Compute continuous and undersampled signals
x_continuous = continuous_signal(t_continuous, f0)
x_undersampled = undersampled_signal(n_undersampled, f0, fs)

# Plotting
plt.figure(figsize=(10, 6))

# Plot continuous signal
plt.plot(t_continuous, x_continuous, label=r'$x(t) = \cos(2\pi f_0 t)$', color='blue')

# Plot undersampled signal
plt.stem(n_undersampled, x_undersampled, label=r'$x[n]$', basefmt='r-', linefmt='r-', markerfmt='ro')

plt.title('Effect of Undersampling')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
