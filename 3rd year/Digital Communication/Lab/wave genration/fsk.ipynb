```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd

m = []
m_len = 8

k = [0,1]
for i in range(m_len):
    m.append(rd.choice(k))

bfsk = []

for i in m:
    if i == 1:
        for i in range(100):
            bfsk.append(1)
    else:
        for i in range(100):
            bfsk.append(0)

t = np.arange(0,m_len,0.01)
bfsk_wave = np.array(bfsk)*np.cos(2*np.pi*2*t)
plt.plot(t,bfsk_wave)
plt.plot(t,bfsk)
plt.grid()
plt.axhline(0,color='red')
plt.title("BPSK Modulated Signal")
plt.xlabel("Time t")
plt.ylabel("Amplitude")
```
