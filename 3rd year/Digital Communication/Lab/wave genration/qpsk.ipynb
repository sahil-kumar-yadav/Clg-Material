```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
#create bit stream 

m = []
m_len = 12

k = [0,1]
for i in range(m_len):
    m.append(rd.choice(k))
```

```python
# splitting message into odd-even bit streams

odd = []
even = []
for i in range(m_len):
    if i%2 == 0: # since python natural indexing start from 0,1
        odd.append(m[i]) #even indices are appended to odd arr
    else:
        even.append(m[i])
```

```python
o_bpsk = []
e_bpsk = []
for i in odd:
    if i==1:
        for j in range(100):
            o_bpsk.append(1)
    else:
        for j in range(100):
            o_bpsk.append(-1)

for k in even:
    if k == 1:
        for j in range(100):
            e_bpsk.append(1)
    else:
        for j in range(100):
            e_bpsk.append(-1)
```

```python
t = np.arange(0,6,0.01)
qpsk = np.array(e_bpsk)*np.sin(2*np.pi*2*t) + np.array(o_bpsk)*np.cos(2*np.pi*2*t)
plt.plot(t,qpsk)
plt.xticks(np.arange(0,6,1))
plt.grid()
plt.axhline(0,color='red')
plt.title("Qpsk modulated wave")
plt.xlabel("Time t")
plt.ylabel("Amplitude")
for i in range(6):
    plt.axvline(i,linestyle='-.',color='green',alpha=0.5)
```
