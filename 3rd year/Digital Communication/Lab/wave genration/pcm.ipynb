```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import signal
pi = np.pi
```

```python
fm = 1 #1 eval(input ('Signal frequency'))
fs = 20 # eval(input('sampling frequency'))
t = np.arange(0,0.001+2/fm,.001)
ts = np.arange(0,1/fs+2/fm,1/fs)
print(len(t))
print(len(ts))
Amax = 1
y = Amax*np.sin(2*pi*fm*t)
ys = Amax*np.sin(2*pi*fm*ts)
plt.stem(ts,ys,'r')
plt.plot(t,y)

plt.show
L= 8 #eval(input('no. of quantization level'))
N = math.log2(L)
N = int(N)
delta = 2*Amax/L
```

```python
levels = np.arange(-Amax,Amax+delta,delta) # Generates an array levels representing the quantization levels from -Amax to Amax with step size delta
code = np.arange(-Amax+delta/2,Amax,delta) #  Generates an array code representing the midpoints of each quantization interval
print(levels)
print(code)
m = len(levels) #  Assigns the number of quantization levels to m
b = m - 1 # alculates the index of the maximum quantization level
q = len(ts)
print(q)
x_quant = np.zeros(q)
x_quant_stair = np.zeros(len(t))
print(len(x_quant_stair))
for num1 in range(q):
    for num2 in range(1,m):
        if ys[num1]==Amax:
            x_quant[num1]=(code[b-1])
        elif ys[num1]>=levels[num2-1] and ys[num1]<levels[num2]:
            x_quant[num1]=(code[num2-1])
x_quant_stair=np.repeat(x_quant[0:len(x_quant)-1],50)
x_quant_stair=np.append(x_quant_stair,x_quant[len(x_quant)-1])
print(len(x_quant_stair))
print(x_quant[len(x_quant)-1])
```

```python
#print(ys)
#print(x_quant)
plt.plot(t,y)
plt.step(t,x_quant_stair)
plt.show 
len_code=len(code)
dec_number=np.zeros(len_code)
for i in range(len_code):
 dec_number[i]=i
def decimalToBinary(n):
 return"{0:b}".format(int(n))
dec_bin=[]
for i in range(len_code):
 dec_bin.append(decimalToBinary(dec_number[i]))
 
print(dec_bin)
```

```python
for i in range(len_code):
 var=dec_bin[i]
 if len(var)<N:
    for j in range(N-len(var)):
        var='0'+var
        dec_bin[i]=var
 
print(dec_bin) 
```
