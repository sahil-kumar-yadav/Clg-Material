```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft
pi = np.pi

t = np.arange(0,1,0.01)
y = np.sin(2*pi*1*t)

ts = 10
ys = []

for i in range(len(y)):
    if(i%10 == 0):
        ys.append(y[i])
    else:
        ys.append(0)

y_dft = fft(ys)

# plt.plot(abs(y_dft[2:100]))
plt.plot(t,y)
plt.stem(t,ys,'r')
# plt.grid()
# plt.axhline()
```

```python
from scipy.signal import butter, filtfilt
from scipy import signal
# reconstruction

y_re = []
prev = 0
for i in range(len(ys)):
    if(ys[i] == 0):
        y_re.append(prev)
    else:
        y_re.append(y[i])
        prev = y[i]

# Smooth the staircase-like output using a moving average filter
window_size = 10  # Adjust window size as needed
y_smooth = np.convolve(y_re, np.ones(window_size)/window_size, mode='same')


plt.plot(t,y_re)
plt.plot(t,y_smooth)
plt.axhline()
```
