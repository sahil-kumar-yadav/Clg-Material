```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd

#create bit stream 

m = []
m_len = 8
# d = np.arange(0,m_len)
k = [0,1]
for i in range(m_len):
    m.append(rd.choice(k))
```

```python
bpsk = []
for i in m:
    if i == 1:
        for j in range(100):
            bpsk.append(1)
    else:
        for j in range(100):
            bpsk.append(-1)
```

```python
t = np.arange(0,m_len,0.01)
bpsk_wave = np.array(bpsk)*np.cos(2*np.pi*2*t)
plt.plot(t,bpsk_wave)
plt.plot(t,bpsk)
# plt.plot(t,bpsk,linewidth=3)
# plt.grid()
# plt.axhline(0,color='red')
plt.title("BPSK Modulated Signal")
plt.xlabel("Time t")
plt.ylabel("Amplitude")
```
